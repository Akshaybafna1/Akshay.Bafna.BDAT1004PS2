# Q1

```python
a = 0

def b():
    global a
    a = c(a)

def c(a):
    return a + 2

b()
b()
b()

print(a)
```

a = 0: Initializes a global variable a with a value of 0.

def b(): Defines a function b().

global a: Inside the b() function, declares that a is a global variable.

a = c(a): Calls the c() function with the current global value of a, adds 2 to it, and assigns the result back to the global variable a.

def c(a): Defines a function c(a) which returns a + 2.

Repeating steps 4-5 three times by calling b() three times:

First call: a becomes 2.
Second call: a becomes 4.
Third call: a becomes 6.
print(a): Outputs the final value of the global variable a, which is 6.

```python
class Marsupial:
    def __init__(self):
        self.myLst = []
        
    def put_in_pouch(self,item):       
        self.myLst.append(item)
        
    def pouch_contents(self):          
        return self.myLst

class Kangaroo(Marsupial):
    def __init__(self,x,y):        
        Marsupial.__init__(self)
        self.x = x
        self.y = y
        
    def jump(self,dx,dy):          
        self.x += dx
        self.y += dy
        
    def __str__(self):            
        return 'I am a Kangaroo located at co-ordinates ({},{})'.format(self.x,self.y)
    
    
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents()
a = Kangaroo(0,0)
print(a)
a.put_in_pouch('doll')
a.put_in_pouch('firetruck')
a.put_in_pouch('kitten')
ans=a.pouch_contents()
print(ans)


a.jump(1,0)
a.jump(1,0)
a.jump(1,0)

print(a)
```

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    Hd = ["h1", "h2"]
    cur = -1

    def handle_starttag(self, tag, attrs):      
        if tag in HeadingParser.Hd:         
            HeadingParser.cur = HeadingParser.Hd.index(tag)
    
    def handle_endtag(self, tag):                
        if tag == HeadingParser.Hd[HeadingParser.cur]:          
            HeadingParser.cur = -1
            
    def handle_data(self, data):
        if HeadingParser.cur >= 0:
            print(" " * HeadingParser.cur + data)
            
infile = open("w3c.html","r")
content = infile.read()
infile.close()
q6 = HeadingParser()
q6.feed(content)
```
